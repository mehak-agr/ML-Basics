```python
df1=df[df.stroke==1]
df1=df1[df1.age>40].sample(500)
df0=df[df.stroke==0]
df0=df0[df0.age>40].sample(500)
```

```python
plt.hist(df1.avg_glucose_level,histtype='step',label='Yes')
plt.hist(df0.avg_glucose_level,histtype='step',label='No')
plt.legend(loc='upper right')
plt.show()
```

```python
plt.hist(df1.age,histtype='step',label='Yes')
plt.hist(df0.age,histtype='step',label='No')
plt.legend(loc='upper right')
plt.show()
```

```python
plt.hist(df1.bmi,histtype='step',label='Yes')
plt.hist(df0.bmi,histtype='step',label='No')
plt.legend(loc='upper right')
plt.show()
```

```python
bins=np.linspace(40,100,10)
plt.hist(df1.age,bins,histtype='step',label='Yes')
plt.hist(df0.age,bins,histtype='step',label='No')
plt.legend(loc='upper right')
plt.show()
```

```python
bins=np.linspace(40,100,10)
plt.hist(df1.age,bins,cumulative=True,histtype='step',label='Yes')
plt.hist(df0.age,bins,cumulative=True,histtype='step',label='No')
plt.legend(loc='upper right')
plt.show()
```

We can conclude age is an important factor in stroke probability determination. Age and stroke have direct dependency.

```python
bins=np.linspace(0,60,30)
plt.hist(df1.bmi,bins,cumulative=False,histtype='step',label='Yes')
plt.hist(df0.bmi,bins,cumulative=False,histtype='step',label='No')
plt.legend(loc='upper right')
plt.show()
```

```python
bins=np.linspace(0,60,30)
plt.hist(df1.bmi,bins,cumulative=True,histtype='step',label='Yes')
plt.hist(df0.bmi,bins,cumulative=True,histtype='step',label='No')
plt.legend(loc='upper right')
plt.show()
```

We can probably conclude, BMI and stroke probability have inverse relationship
