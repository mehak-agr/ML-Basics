Concatenation

```python
india=pd.DataFrame({'City':['Mumbai', 'Goa', 'Delhi'],
                       'Temperature':[32,28,30],
                       'Humidity':[80,90,40]
                      })
india
us=pd.DataFrame({'City':['New York', 'Chicago', 'Orlando'],
                       'Temperature':[30,25,14],
                       'Humidity':[60,70,44],
                      })
us
canada=pd.DataFrame({'City':['Winnipeg', 'Toronto', 'Montreal'],
                       'Temperature':[56.7,14.5,20],
                       'Humidity':[40,56,50],
                      })
canada
```

```python
df=pd.concat([india,us,canada])
df
```

```python
df=pd.concat([india,us,canada],ignore_index=True)
df
```

```python
df=pd.concat([india,us,canada],keys=['India','US'])
df
```

```python
df.loc['US']
```

```python
df.loc['India']
```

Replacing Missing Values

```python
df=pd.read_csv("weather_data.csv")
df
```

```python
temp_df=df.replace(to_replace=['Sunny'],value=-999)
temp_df
```

```python
temp_df=df.replace(to_replace=np.nan,value=-999)
temp_df
```

```python
temp_df=df.replace('Rain',value=np.NaN)
temp_df
```

```python
temp_df=df.replace([32,28],value=0)
temp_df
```

```python
temp_df=df.replace({'temperature':32,'windspeed':7,'event':'Snow'},np.nan)
temp_df
```

When windspeed is 6mph, 7mph etc and temperature is 32F, 28F etc 
then regex will remove that mph and F and other such strings.

```python
temp_df=df.replace({'temperature':'[A-Za-z]','windspeed':'[a-z]'},'',regex=True)
temp_df
```

Merge

```python
df1=pd.DataFrame({'city':['new york','chicago','orlando'],
                  'temperature':[21,14,35]
                 })
df1
df2=pd.DataFrame({'city':['chicago','new york','orlando'],
                  'humidity':[65,63,56]
                 })
df2
df=pd.merge(df1,df2,on='city')
df
```

```python
df1=pd.DataFrame({'city':['new york','chicago','orlando','baltimore'],
                 'temperature':[45,34,67,89]                
                })
df1
df2=pd.DataFrame({'city':['chicago','new york','san diego'],
                  'humidity':[34,67,98]
                 })
df2
df=pd.merge(df1,df2,on='city',how='inner')
df
df=pd.merge(df1,df2,on='city',how='outer')
df
df=pd.merge(df1,df2,on='city',how='left')
df
df=pd.merge(df1,df2,on='city',how='right')
df
df=pd.merge(df1,df2,on='city',how='outer',indicator=True)
df
```

CrossTab

```python
df=pd.read_excel("hand_data.xlsx",header=[1])
df
```

```python
a=pd.crosstab(df.Nationality,df.Handedness)
a
b=pd.crosstab(df.Sex,df.Handedness)
b
c=pd.crosstab(df.Sex,df.Handedness,margins=True)
c
```

```python
d=pd.crosstab(df.Sex,[df.Handedness,df.Nationality],margins=True)
d
e=pd.crosstab([df.Nationality,df.Sex],[df.Handedness],margins=True)
e
f=pd.crosstab(df.Sex,df.Handedness,normalize='index')
f
```

```python
g=pd.crosstab(df.Sex,df.Handedness,values=df.Age,aggfunc=np.average)
g
```
