```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv('DTree.csv',skiprows=0,header=1)
df.head()
```

```python
#divide it into target and independent variable
inputs=df.drop('Salary_more than 100k',axis='columns')
target=df['Salary_more than 100k']
inputs
target
```

```python
#create columns into numbers because ML works on numbers
from sklearn.preprocessing import LabelEncoder
le_company=LabelEncoder()
le_job=LabelEncoder()
le_degree=LabelEncoder()
```

```python
inputs['Company_n']=le_company.fit_transform(inputs['Company'])
inputs['Job_n']=le_job.fit_transform(inputs['Job'])
inputs['Degree_n']=le_degree.fit_transform(inputs['Degree'])
inputs.head()
```

```python
inputs_n=inputs.drop(['Company','Job','Degree'],axis='columns')
inputs_n
```

```python
#train classifer
from sklearn import tree
model=tree.DecisionTreeClassifier()
model.fit(inputs_n,target)
model.score(inputs_n,target)
model.predict([[2,2,1]])
```
