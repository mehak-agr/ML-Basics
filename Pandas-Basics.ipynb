```python
df
#Find and Replace analogy to Word
def convert_people_cell(cell):
    if cell=="n.a.":
        return 'Sam Wall'
    return cell
def convert_price_cell(cell):
    if cell=="n.a.":
        return 50
    return cell
df=pd.read_excel("stock_data.xlsx",
                 converters={'people':convert_people_cell,'price':convert_price_cell})
df
df.to_excel("new.xlsx",sheet_name="stocks",index=False,startrow=2,startcol=1)
df
```

```python
df_stocks=pd.DataFrame({'tickers':['GOOGLE','WMT','MSFT'],
                        'price':[845,65,64],
                        'pe':[30.37,14.26,30.97],
                        'eps':[27.82,4.61,2.12]
                       })
df_weather=pd.DataFrame({'day':['1/1/2017','1/2/2017','1/3/2017'],
                         'temperature':[32,35,28],
                         'event':['Rain','Snow','Sunny']
                        })
with pd.ExcelWriter('stocks_weather.xlsx') as writer:
    df_stocks.to_excel(writer, sheet_name="stocks")
    df_weather.to_excel(writer, sheet_name="weather")
df=pd.read_excel("stocks_weather.xlsx","weather")
df
```
